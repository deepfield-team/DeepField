### Basic show method

Required imports

```python
import sys
sys.path.append('..')
from deepfield import Field
```

Let's load some reservoir model, for example, the `Egg` model provided in the `open_data` directory (remember to unzip the file `RESULTS.zip` in the `open_data/egg` directory):

```python
model = Field('../open_data/egg/Egg_Model_ECL.DATA').load()
```

Run the method `show` indicating some grid cell parameter, for example, the x-permeability (`PERMX`)

```python
model.show(attr='PERMX', slicing=True, cmap='jet')
```

This will open a separate interactive window that shows the 3D reservoir model, the distribution of the specified parameter over grid cells, well trajectories and their names. If the model contains faults, they will also be shown. Use the mouse to rotate the reservoir model and scroll to zoom in and out. By default, we use the standard `matplotlib` color map and half opacity for grid cells.

<p>

Move the opacity slider to make the model more or less transparent. Unclick the button `Wells` to hide well trajectories and well names. The `Egg` model has no faults, so the `Faults` button has no effect.

<p>

### Colormap

Use the parameter `cmap` to choose any colormap available in `matplotlib`. See [https://matplotlib.org/stable/users/explain/colors/colormaps.html](https://matplotlib.org/stable/users/explain/colors/colormaps.html) for the list of colormap names. In this example, we set `cmap='jet'`: 

```python
model.show(attr='PERMX', cmap='jet')
```

<p>

### Slicing

The parameter `slicing=True` shows the reservoir model by slices. The three sliders that appear allow you to select slices in each direction (x, y and z):

```python
model.show(attr='PERMX', slicing=True, cmap='jet')
```

<p>

### Thresholding

You can hide cells with values below some threshold by setting `thresholding=True`. This will add a slider that allows you to control the lower boundary of the values:

```python
model.show(attr='PERMX', thresholding=True, cmap='jet')
```

<p>

### Theme

Prefer dark modes? Try `theme='ParaView'` or `theme='dark'` to switch to darker background. In this example, we use `theme='ParaView'`:

```python
model.show(attr='PERMX', theme='ParaView', cmap='jet')
```

<p>

### Dynamic attributes

The arrays in the `States` component have a time dimension. Move the `Timestamp` slider to see how the values change over time. Let's show the `PRESSURE` field for an example:

```python
model.show(attr='PRESSURE', theme='ParaView', cmap='jet')
```

<p>

### Cell edges

Note that by default we show edges of each grid cell. When it is not necessary, set `show_edges=False`:

```python
model.show(attr='PRESSURE', theme='ParaView', cmap='jet', show_edges=False)
```

<p>

### Scaling

By default, we scale each dimension by its extent. As a result, the model looks as if it fits a unit cube. However, in reality, the original scales can be very different. You can see the model in its original proportions by setting `scaling=None`:

```python
model.show(attr='PRESSURE', theme='ParaView', cmap='jet', show_edges=False, scaling=None)
```

<p>

You can apply custom scaling to change the proportions of the model. For example, below we use `scaling=(1, 2, 20)`:

```python
model.show(attr='PRESSURE', theme='ParaView', cmap='jet', show_edges=False, scaling=(1, 2, 20))
```

<p>
